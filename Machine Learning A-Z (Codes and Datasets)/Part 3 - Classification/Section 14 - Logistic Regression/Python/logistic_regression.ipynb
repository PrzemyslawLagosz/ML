# Logistic Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
dataset.head()
```

```python
dataset.describe()
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
X_test
```

## Training the Logistic Regression model on the Training set

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression(random_state= 0)
classifier.fit(X_train, y_train)
```

## Predicting a new result

### For castomer `30 age` and earingin `87k`

```python
scaled_new_customer = sc.transform([[30, 87000]])
scaled_new_customer
```

```python
classifier.predict( scaled_new_customer )
```

## Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
y_test
```

```python
compare_df = pd.DataFrame({"y_pred": classifier.predict(X_test), "y_test": y_test})
compare_df.head()
```

### Tyle procent dobrze przewidział model

```python
(compare_df.y_pred == compare_df.y_test).sum() / len(compare_df)
```

## Making the Confusion Matrix

* 65 przewidziane NIE kupił, prawda: NIE kupił
* 8 przewidziane kupił, prawda: nie kupił
* 3 przewidziane NIE kupić prawda: kupił
* 24 przewidziane kupił, prawda: kupił


```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

print(cm)
```

```python
from sklearn.metrics import accuracy_score

acc_score = accuracy_score(y_test, y_pred)

print(acc_score)
```

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.5),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.5))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('salmon', 'dodgerblue')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('salmon', 'dodgerblue'))(i), label = j)
plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
